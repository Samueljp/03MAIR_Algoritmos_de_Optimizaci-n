Respuesta:

Con restriccin de un partido por en cada horario: 
- Dado que hay 10 partidos y 10 horarios lo ms lgico sera responder que eixten 10x9! posibilidades, por lo tanto N x (N-1)!. Esta solucin ocupara todos los horarios, por lo cual el lunes y el viernes. Con las condiciones del problema 

Sin restriccin de horarios:
- Si tenemos en cuenta que se podran acumular varios partidos en el mismo horario siempre que el viernes y el lunes tengan un partido, por ejemplo, 9 partidos el viernes y 1 el lunes, 8 partidos el viernes, 1 el lunes y cualquier otro horario el restante y as continuar... 
En este caso tendramos(Siendo N el nmero de partidos y r el nmero de partidos agrupados):
$\sum_{r=1}^{N-1} \frac{N!}{ r!  (N-r)!} (\frac{(N-1)!  ((N-1)-r)!}{ r!  ((N-1)-r)!} +(N-r)!)$

Calculado basndome en la funcin de itertools de Python ms un bucle que he realizado

Complejidad de combinations:
https://docs.python.org/3/library/itertools.html#itertools.combinations

Complejidad de permutations:
https://docs.python.org/3/library/itertools.html#itertools.permutations
 

```python
import math
import itertools as it

N = 10
count = 0

# Posibles agrupamientos
for r in range(N-1):
  aux = 0
  ## Posibles casos de agrupacin del partido del viernes
  for x in it.combinations(range(1,N),(N-r)):
    x = N-1
    ## Posibles casos de agrupacin del partido del lunes con los restantes del viernes
    for y in it.combinations(range(1,x),(x-r)):
      # Distintas colacaciones en los horarios resultantes
      aux += math.factorial(x-r)
  # Combinativa con posibles partidos entre los horarios restantes
  # Distintas colocaciones en los horarios resultantes
  aux += math.factorial(N-r)
  count += aux
  

print("Iteraciones: ",count)

print("Ejemplo de casuistica de partidos (3 partidos de 5 totales) en un horario")
[ (set(x), (set([j for j in range(5)]))-set(x)) for x in it.combinations(range(5), 3)]
```

La funcin objetivo es por ejemplo la funcin en la cual ya asignados los partidos a los horarios, calcula la audiencia que tendr la jornada sumando la audiencia de cada uno de los partidos.

Es un problema de maximizacin, ya que se quiere la mayor audiencia por jornada.

```python
# Calcula la audiencia de una jornada
# JORNADA: Estructura con la planificacin de la jornada
# COEF_JORNADA: Coeficientes multiplicadores dependiendo del horario (da y hora)
# AUDIENCIA_CATEGORIA: Audiencia que generan los partidos debido a la categora de los equipos
# PUNTOS_CATEGORIA: Puntos de reparto de audiencia segn categora de equipos
# con_resumen: Parmetro opcional que implica la devolucin de un resumen de la puntuacin calculada
# RETURN: Number -> Audiencia de JORNADA
def calculaAudienciaJornada(JORNADA, COEF_HORARIOS, AUDIENCIA_CATEGORIA, PUNTOS_CATEGORIA, con_resumen=False):
    
    audiencia_total = 0
    resumen_jornada = []
    # Recorrido de los partidos de la jornada por los horarios
    for horario, partidos in JORNADA:
        
        # Obtencin de mxima audiencia para horario
        coef_max_audiencia = 0        
        for partido in partidos:
            
            #print("Equipo1: %s, Equipo2: %s " % (partido[0][1],partido[1][1]))
            if coef_max_audiencia < PUNTOS_CATEGORIA[partido[0][1]] + PUNTOS_CATEGORIA[partido[1][1]]:
                coef_max_audiencia = PUNTOS_CATEGORIA[partido[0][1]] + PUNTOS_CATEGORIA[partido[1][1]]
#             else :
#                 coef_max_audiencia += PUNTOS_CATEGORIA[partido[0][1]] + PUNTOS_CATEGORIA[partido[1][1]]
        
        # Calculo de audiencia por partido
        for partido in partidos:
        
            audiencia = AUDIENCIA_CATEGORIA[ str(partido[0][1] + partido[1][1]) ]
            coef_horario = COEF_HORARIOS[horario]
            coef_audiencia = PUNTOS_CATEGORIA[partido[0][1]] + PUNTOS_CATEGORIA[partido[1][1]]
            # Audiencia de equipos * coef_horario * ponderacin comparticin
            audiencia_partido = audiencia * coef_horario *(coef_audiencia / coef_max_audiencia)
            audiencia_total += audiencia_partido
            
            # Generacin de texto con resumen de los clculos
            if con_resumen:
                resumen = {"audiencia": audiencia, 
                           "coef_horario": coef_horario, 
                           "coef_audiencia(%)": str(coef_audiencia) +"/" + str(coef_max_audiencia),
                           "audiencia_partido": audiencia_partido
                          }
                resumen_jornada.append({"horario":horario,
                                        "partido":partido,
                                        "resumen":resumen})
            # Fin de generacin de texto con resumen
           
    if con_resumen: return (audiencia_total,resumen_jornada)    
    return audiencia_total
```

```python
# Importacin de librera para obtener todas las posibles combinaciones
import itertools as it

# Funcin que calcula el mejor horario por fuerza bruta
# Partidos: Jornada con los enfrentamientos entre equipos (Partidos)
# COEF_JORNADA: Coeficientes multiplicadores dependiendo del horario (da y hora)
# AUDIENCIA_CATEGORIA: Audiencia que generan los partidos debido a la categora de los equipos
# PUNTOS_CATEGORIA: Puntos de reparto de audiencia segn categora de equipos
# Return: La mejor reparticin de los partidos a travs de la jornada
def calcularJornadaFB(PARTIDOS,COEF_HORARIOS,AUDIENCIA_CATEGORIA,PUNTOS_CATEGORIA):
    
    # Lista de horarios
    lstHorarios = [ x for x in COEF_HORARIOS]
    
    # Posibles combinaciones N!
    combinaciones = it.permutations(PARTIDOS['Equipos'],len(PARTIDOS['Equipos']))
    
    i = 0
    mejor_jornada = []
    mejor_puntuacion = 0
    # Bucle que recorre los posibles casos
    for c in combinaciones:
        i+=1
        
        # Generacin de jornada en funcin de permutacin
        jornada = []
        horario = 0
        
        # Bucle para asignacin de los partidos a un horario
        for partido in c:
            
            jornada.append((lstHorarios[horario],[partido]))
            horario +=1
        
        # Calculo de puntuacin de Jornada
        puntuacion = calculaAudienciaJornada(jornada,COEF_HORARIOS,AUDIENCIA_CATEGORIA,PUNTOS_CATEGORIA)
        
        # Permanencia de mejor Jornada
        if puntuacion > mejor_puntuacion:
            mejor_jornada = jornada
            mejor_puntuacion = puntuacion
        
#     print("Combinaciones: %d" %i)
    return (mejor_jornada,mejor_puntuacion)
```

```python
# https://es.wikipedia.org/wiki/Sistema_de_todos_contra_todos
# from random import shuffle    

# Generador de calendario Estandar
def getCalendarioClubes(clubes): 

    index_clubes = 0    
    auxT = len(clubes)

    totalP = auxT//2 # total de partidos de una jornada
    jornada = []
    indiceInverso = auxT-2

    for i in range(1,auxT):
        equipos = []
        list_equipos = {}
        
        for indiceP in range(0,totalP):
            
            if index_clubes > auxT-2: index_clubes = 0
    
            if indiceInverso < 0: indiceInverso = auxT-2
    
            if indiceP == 0: # seria el partido inicial de cada fecha
                partido = [clubes[index_clubes], clubes[auxT-1]]
#                 print("index_clubes: %d, auxT-1: %d" % (index_clubes,(auxT-1)))
                equipos.append(partido)
            else:
                partido = [clubes[index_clubes], clubes[indiceInverso]]
#                 print("index_clubes: %d, indiceInverso: %d" % (index_clubes,indiceInverso))
                equipos.append(partido)
                indiceInverso -= 1
            index_clubes += 1

        list_equipos = { 'Jornada': "Jornada Nro.: " + str(i),
                        'Equipos': equipos}
        jornada.append(list_equipos)

    return jornada
    
calendario = getCalendarioClubes(equiposM)
calendario
```
